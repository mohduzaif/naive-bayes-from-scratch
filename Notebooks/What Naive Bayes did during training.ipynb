```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('../Datasets/play_tennis.csv')
```

```python
df
```

```python
# these are the joint probability 
pd.crosstab(df['outlook'], df['play'], normalize='all')
```

```python
# these are the marginal and joint probability 
pd.crosstab(df['outlook'], df['play'], normalize='all', margins=True)
```

```python
# these are the conditional probability 
pd.crosstab(df['outlook'], df['play'], normalize='columns')
```

```python
# how it calculate the probabilites and save them as dictionary. 
pd.crosstab(df['outlook'], df['play'], normalize='columns').stack().to_dict()
```

```python
# for all the columns/features. 
final_prob = {} 
for column in ['outlook', 'temp', 'humidity', 'wind']: 
    final_prob.update(pd.crosstab(df[column], df['play'], normalize='columns').stack().to_dict())
```
