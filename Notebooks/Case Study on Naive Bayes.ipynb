### Text Cleaning. 

1. Sample 10000 rows
2. Remove html tags
3. Remove special characters
4. Converting every thing to lower case
5. Removing Stop words
6. Stemming

```python
review_df = review_df.sample(1000)
```

```python
# make 0 and 1 in the output columns(sentiment). 
review_df['sentiment'].replace({'positive' : 1, 'negative' : 0}, inplace = True)
```

```python
review_df.head()
```

```python
import re
clean = re.compile('<.*?>')
re.sub(clean, '', review_df.iloc[2].review)
```

```python
# Function to clean html tags
def clean_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
review_df['review']=review_df['review'].apply(clean_html)
```

```python
# converting everything to lower

def convert_lower(text):
    return text.lower()
```

```python
review_df['review']=review_df['review'].apply(convert_lower)
```

```python
# function to remove special characters

def remove_special(text):
    x=''
    
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x + ' '
    return x
```

```python
remove_special(' th%e @ classic use of the word.it is called oz as that is the nickname given to the oswald maximum security state penitentary. it focuses mainly on emerald city, an experimental section of the prison where all the cells have glass fronts and face inwards, so privacy is not high on the agenda. em city is home to many..aryans, muslims, gangstas, latinos, christians, italians, irish and more....so scuffles, death stares, dodgy dealings and shady agreements are never far away.i would say the main appeal of the show is due to the fact that it goes where other shows wouldnt dare. forget pretty pictures painted for mainstream audiences, f')
```

```python
review_df['review']=review_df['review'].apply(remove_special)
```

```python
# Remove the stop words
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english')) 
stop_words
```

```python
# function to remove the stop words. 
def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)
```

```python
review_df['review'] = review_df["review"].apply(remove_stopwords)
```

```python
review_df.head()
```

```python
# Perform stemming

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer() 
```

```python
y=[]
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
stem_words(['I','loved','loving','it'])
```

```python
# apply the stemwords. 
review_df['review']=review_df['review'].apply(stem_words)
```

```python
review_df
```

```python
# Join back
def join_back(list_input):
    return " ".join(list_input)
```

```python
review_df['review']=review_df['review'].apply(join_back)
```
